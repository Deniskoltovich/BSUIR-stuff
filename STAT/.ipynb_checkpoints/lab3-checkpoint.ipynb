```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
PATH = "./lab1/data.csv"
data = pd.read_csv(PATH)

data = data.drop(["Unnamed: 0", "Description","Stars", 'Director', 'Movie Name'], axis=1)
display(data.head())

label_encoder = LabelEncoder()
data["Genre"] = label_encoder.fit_transform(data["Genre"])
data["Certification"] = label_encoder.fit_transform(data["Certification"])

# Заполнение отсутствующих значений средним по столбцу
imputer = SimpleImputer(strategy="mean")
data["MetaScore"] = imputer.fit_transform(data[["MetaScore"]])
data["Gross"] = imputer.fit_transform(data[["Gross"]])

# Масштабирование числовых признаков
scaler = StandardScaler()
numerical_features = ["Year of Release", "Run Time in minutes", "Movie Rating", "Votes", "MetaScore", "Gross"]
data[numerical_features] = scaler.fit_transform(data[numerical_features])
data
```

```python
inertia = []
for i in range(1, 11):
    k_means = KMeans(n_clusters=i, init= 'k-means++')
    k_means.fit(data)
    inertia.append(k_means.inertia_)
    
sns.set_style('darkgrid')
sns.scatterplot(
x=[x for x in range(1, 11)],
y=inertia,
)
plt.title('График зависимости')
plt.xlabel('Количество кластеров')
plt.ylabel('Внутри-кластерная сумма расстояний')
```

```python
CLUSTERS = 4
model = KMeans(n_clusters=CLUSTERS)
model.fit(data)
# Модель пытается разделить данные на указанное количество кластеров

clusters = pd.DataFrame(
    columns=data.columns,
    data=model.cluster_centers_
    )
clusters["Amount"] = np.unique(
model.labels_,
return_counts=True
)[1]
clusters
```
